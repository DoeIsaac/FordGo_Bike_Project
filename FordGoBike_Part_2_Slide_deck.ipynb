```python
# import all packages and set plots to be embedded inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline

# suppress warnings from final output
import warnings
warnings.simplefilter("ignore")
```

```python
# Load dataset
data = pd.read_csv('updated_fordgobike.csv')
data
```

## visualization
After data wrangling, we analyze many features to gather insight and answer few question. 

We found out the Gender and Age group that need more promotion and advertisement. Majority of the population in our dataset are Male while the remeaning few are mostly Female with little percentage of other. We need to look into the transport needs of female and other, and how we can introduce new campaign to promote our services to Females and others.


Meanwhile, most of our users are youth and young adults, a little percentage are Elderly and Adults. Meaning, We could choose to improve our services for the young people or introduce self driven and automatic bikes to attract more older people to our services.

```python
# Lets order the age group from young to older
order_age =['Youth','Young Adult','Adult','Elderly']
ordered_age_group = pd.api.types.CategoricalDtype(ordered = True, categories = order_age)
data['age_group'] = data['age_group'].astype(ordered_age_group)

# Ploting the necessary chart
sns.countplot(data= data, x="age_group", hue="member_gender")
plt.title('gender per age group')
plt.legend(loc = 'upper right');
```

The analysis also shows that younger people have the tendency to travel longer than older people, it could be because of their risk tolerance, strength and eagernesss to migrate frequently etc. 


```python
sns.scatterplot(data=data, x='member_birth_year', y='duration_sec', hue='user_type')
plt.xlabel('Birth year')
plt.ylabel('duration(sec)')
plt.title('Relationship Between birth year and duration')
plt.savefig('age_duration.png');
```

Seeing around 90% of the population in each age group are subscriber and few customers is awesome. This shows we are doing a good job and many users have subscribed to our services. We now need to find a way to make subscribers return often. We could decide to offer discount monthly to subscribers only, this could the few customers to subscribe also.


```python
sns.countplot(data= data, x="member_gender", hue="user_type")
plt.title('user_type per gender')
plt.legend(loc = 'upper right')
plt.savefig('user_type.png');
```
